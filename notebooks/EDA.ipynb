```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("india_crop_recommendation_dataset_extended.csv")

df.head()
```

```python
plt.figure(figsize=(12,5))
sns.countplot(x="label", data=df)
plt.xticks(rotation=45)
plt.title("Crop Distribution")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.histplot(df["N"], kde=True)
plt.title("Distribution of Nitrogen")
plt.show()
```

```python
df[['N','P','K','temperature','humidity','ph','rainfall']].hist(
    figsize=(14,10), bins=20
)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.drop("label", axis=1).corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(x="label", y="rainfall", data=df)
plt.xticks(rotation=45)
plt.title("Rainfall vs Crop Type")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(x="label", y="temperature", data=df)
plt.xticks(rotation=45)
plt.title("Temperature vs Crop")
plt.show()
```
